```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = [1,2,3,4]
error_estimate_1 = []
error_estimate_2 = []

for t in theta:
    y = np.random.uniform(0,t,10000)
    estimate1 = max(y)
    estimate2 = 2*np.mean(y)

    error_estimate_1.append(np.abs(estimate1 - t))
    error_estimate_2.append(np.abs(estimate2 - t))

fig, ax = plt.subplots(1, 1, figsize=(10, 6))

ax.plot(theta, error_estimate_1, marker='o', linewidth=2, markersize=8, label='Estimate 1: Max')
ax.plot(theta, error_estimate_2, marker='s', linewidth=2, markersize=8, label='Estimate 2: 2*Mean')

ax.set_xlabel('Parameter(theta)', fontsize=12)
ax.set_ylabel('Absolute Error', fontsize=12)
ax.set_title('Error Analysis: Max vs 2*Mean Estimates', fontsize=14)
ax.set_xticks(theta)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()
```
