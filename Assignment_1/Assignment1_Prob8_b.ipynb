```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Number_of_samples = [10,50,100,1000,5000,10000]
estimate = []

for N in Number_of_samples:
    y = np.random.uniform(0,1,N)
    estimate1 = np.max(y)
    estimate.append(abs(1 - estimate1))

fig, ax = plt.subplots(1, 1, figsize=(10, 6))

#Plot in semi-log scale
ax.plot(Number_of_samples, estimate, marker='o', linewidth=2, markersize=8, label='Estimate: Max')
ax.set_xscale('log')

ax.set_xlabel('Number of samples(N)', fontsize=12)
ax.set_ylabel('Absolute Error', fontsize=12)
ax.set_title('Estimate vs N', fontsize=14)
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11)

plt.tight_layout()
plt.show()
```
